## Here is a detailed description of each column:

## Gameplay Metrics

**1.Session_Duration:** The length of time a player spends playing the game in a single session.

**2.Daily_Active_Time:** The amount of time a player spends playing the game on a daily basis.

**3.Login_Frequency:** The number of times a player logs into the game within a certain time period (e.g., daily, weekly).

**4.Level_Progression:** The rate at which a player progresses through the game's levels or stages.

**5.In_Game_Currency_Earned:** The amount of in-game currency (e.g., coins, gems) earned by a player.

**6.In_Game_Currency_Spent:** The amount of in-game currency spent by a player on items, upgrades, or other virtual goods.

**7.Achievement_Unlocks:** The number of achievements or milestones unlocked by a player.

**8.Social_Interactions:** The number of social interactions a player has with other players, such as chatting, trading, or teaming up.

**9.Friends_Added:** The number of friends a player has added to their in-game friends list.

**10.Guild_Participation:** The level of participation a player has in a guild or group, such as attending events or contributing to group goals.

## Monetization Metrics

**11.Purchase_History:** A record of a player's past purchases, including the items bought and the amount spent.

**12.In_Game_Purchases:** The number of in-game purchases made by a player.

**13.Favorite_Items:** The items or virtual goods that a player has marked as their favorites.

**14.Discount_Utilization:** The number of times a player has taken advantage of discounts or promotions.

**15.Promotion_Response:** The player's response to promotions, such as clicking on ads or participating in special events.

## Player Behavior Metrics

**16.Game_Mode_Preferences:** The game modes or genres that a player prefers to play.

**17.Event_Participation:** The number of events or tournaments a player has participated in.

**18.Item_Usage:** The frequency and type of items used by a player.

**19.Time_of_Purchase:** The time of day or week when a player makes a purchase.

**20.Retention_Rate:** The percentage of players who continue to play the game over time.

**21.Churn_Rate:** The percentage of players who stop playing the game over time.

**22.Referral_Activity:** The number of referrals or invites sent by a player to their friends
## Player Profile Metrics
**23.Customization_Preferences:** The player's preferences for customizing their game experience, such as character appearance or game settings.

**24.Ad_Engagement:** The player's level of engagement with advertisements, such as clicking on ads or watching videos.

**25.Device_Type:** The type of device used by a player to play the game (e.g., PC, console, mobile).

**26.Geographic_Location:** The player's geographic location, such as country or region.

**27.Demographic_Age:** The player's age or age range.

**28.Demographic_Gender:** The player's gender.

## Player Performance Metrics

**29.Quest_Completion_Rate:** The percentage of quests or missions completed by a player.

**30.Battle_Participation:** The number of battles or PvP matches participated in by a player.

**31.Leaderboard_Position:** The player's ranking on leaderboards or competitive rankings.

**32.Player_Engagement_Level:** A measure of a player's overall engagement and activity level in the game.

**33.Purchase_Likelihood:** A prediction of how likely a player is to make a purchase based on their behavior and preferences.

## ANN

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Convert DataFrame to NumPy array
xshample1 = xtrain.to_numpy()  
yshample1 = ytrain.to_numpy()  

# Splitting the data into training and testing parts
xtrain, xtest, ytrain, ytest = train_test_split(xshample1, yshample1, test_size=0.2, random_state=42)

# Define the ANN model
ANN = Sequential()

# First Dense layer
ANN.add(Dense(128, input_dim=xtrain.shape[1], activation='relu'))

# Second Dense layer
ANN.add(Dense(80, activation='relu'))

# Third Dense layer
ANN.add(Dense(60, activation='relu'))

# Add Dropout to reduce overfitting
ANN.add(Dropout(0.5))

# Add output layer
ANN.add(Dense(1, activation='sigmoid'))  # For binary classification

# Compile the model with a lower learning rate
ANN.compile(optimizer=Adam(learning_rate=0.0005),
            loss='binary_crossentropy',
            metrics=['accuracy'])

# Train the model
history = ANN.fit(xtrain, ytrain, epochs=20, batch_size=32, validation_split=0.2)  # Increased epochs

# Evaluate the model
test_loss, test_accuracy = ANN.evaluate(xtest, ytest)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")



# Evaluate the model
test_loss, test_accuracy = ANN.evaluate(xtest, ytest)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Saving the model in HDF5 format (.h5)
ANN.save('ANN_model.h5')

# Alternatively, saving in TensorFlow's SavedModel format
ANN.save('ANN_saved_model')
```

```python
# Evaluate the model
test_loss, test_accuracy = ANN.evaluate(xtest, ytest)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Saving the model in HDF5 format (.h5)
ANN.save('ANN_model.h5')

# Alternatively, saving in TensorFlow's SavedModel format
ANN.save('ANN_saved_model')
```

```python
from sklearn.metrics import classification_report

# Make predictions
y_pred_prob = ANN.predict(xtest)  # Get probabilities
y_pred = (y_pred_prob > 0.5).astype(int)  # Convert probabilities to binary predictions

# Generate classification report
print("Classification Report:")
print(classification_report(ytest, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelBinarizer

# Assuming 'history' is the output from CNN.fit()
# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'])

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'])

plt.tight_layout()
plt.show()

# Make predictions
y_pred = ANN.predict(xtest)
y_pred_classes = (y_pred > 0.5).astype("int32")

```
